```python
import cv2
import image_dehazer
import matplotlib.pyplot as plt
```

```python
if __name__ == "__main__":

    HazeImg = cv2.imread('/content/drive/MyDrive/ESP32-CAM/20240420/20240420-205959.jpg')            # read input image -- (**must be a color image**)
    HazeCorrectedImg, haze_map = image_dehazer.remove_haze(HazeImg, showHazeTransmissionMap=False)    # Remove Haze

    plt.imshow(HazeCorrectedImg);     # display the result
```

```python
alpha = 1.5
# control brightness by 50
beta = 50
image2 = cv2.convertScaleAbs(HazeImg, alpha=alpha, beta=beta)

#Save the image
cv2.imwrite('Brightness & contrast.jpg', image2)
#Plot the contrast image
plt.subplot(1, 2, 2)
plt.title("Brightness Image")
plt.imshow(image2)
plt.show()
```

```python
rbg_image=cv2.cvtColor(image2,cv2.COLOR_BGR2RGB)
plt.imshow(rbg_image)
```
